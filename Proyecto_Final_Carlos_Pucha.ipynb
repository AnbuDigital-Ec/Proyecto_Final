## LIMPIEZA Y NORMALIZACIÓN

### NORMALIZACIÓN DE CAMPOS CLAVE

```python
df["ACTIVIDAD_ECONOMICA"] = df["ACTIVIDAD_ECONOMICA"].str.upper().str.strip()
df["DESCRIPCION_CANTON_EST"] = df["DESCRIPCION_CANTON_EST"].str.upper().str.strip()
df["DESCRIPCION_PARROQUIA_EST"] = df["DESCRIPCION_PARROQUIA_EST"].str.upper().str.strip()
df["TIPO_CONTRIBUYENTE"] = df["TIPO_CONTRIBUYENTE"].str.upper().str.strip()
df["CLASE_CONTRIBUYENTE"] = df["CLASE_CONTRIBUYENTE"].str.upper().str.strip()

print("Normalización completa...")
```

## VARIABLE ANTIGÜEDAD
### CONVERTIR FECHA A DATETIME

```python
# Convertir FECHA_INICIO_ACTIVIDADES a formato datetime
df["FECHA_INICIO_ACTIVIDADES"] = pd.to_datetime(
    df["FECHA_INICIO_ACTIVIDADES"],
    errors="coerce"  #errors=coerce si existe algun error convierte en NaN
)

# Verificar conversión
df["FECHA_INICIO_ACTIVIDADES"].head()
```

### CREAR ANTIGÜEDAD EN AÑOS

```python
# Fecha actual
fecha_actual = pd.Timestamp.now()

# Calcular antigüedad en años
df["ANTIGUEDAD_ANIOS"] = (
    (fecha_actual - df["FECHA_INICIO_ACTIVIDADES"]).dt.days / 365
)

# Redondear a 1 decimal
df["ANTIGUEDAD_ANIOS"] = df["ANTIGUEDAD_ANIOS"].round(1)

df[["FECHA_INICIO_ACTIVIDADES", "ANTIGUEDAD_ANIOS"]].head()
```

### VER DISTRIBUCIÓN DE ANTIGÜEDAD

```python
df["ANTIGUEDAD_ANIOS"].describe()
```

```python
# Estadísticas descriptivas iniciales de la antigüedad
df["ANTIGUEDAD_ANIOS"].describe()

# Esto nos permite entender el comportamiento general de la variable
# antes de aplicar cualquier filtrado

```

## Calcular IQR y límite superior

```python
# Calcular cuartiles
Q1 = df["ANTIGUEDAD_ANIOS"].quantile(0.25)
Q3 = df["ANTIGUEDAD_ANIOS"].quantile(0.75)

# Calcular rango intercuartílico
IQR = Q3 - Q1

# Calcular límite superior para detectar valores atípicos
limite_superior = Q3 + 1.5 * IQR

limite_superior

# Este valor indica el umbral estadístico a partir del cual
# se consideran outliers según el método IQR

```

## DETECCION Y ELIMINACION DE OUTLIERS
### Identificar outliers

```python
# Identificar registros con antigüedad mayor al límite estadístico
df_outliers = df[df["ANTIGUEDAD_ANIOS"] > limite_superior]

df_outliers.head()

# Estos registros corresponden principalmente a instituciones históricas
# que no representan el comportamiento típico de negocios comerciales
```

### ESTABLECIMIENTOS IDENTIFICADOS CON MAS DE 46.15 AÑOS DE INICIO DE ACTIVIDADES

```python
df_outliers.shape

# Se cuantifica la cantidad de valores atípicos detectados
```

## DATASET FINAL DEPURADO
### FILTRAR ESTABLECIMIENTOS CON MENOS DE 50 AÑOS DE INICIO DE ACTIVIDADES

```python
# Filtrar negocios con antigüedad menor o igual a 50 años
df = df[df["ANTIGUEDAD_ANIOS"] <= 50]

# Se establece 50 años como umbral razonable,
# alineado con el límite estadístico calculado (46 años)
```

### NUEVOS DATOS DESPUÉS DE ELIMINAR VALORES ATIPICOS

## Clasificación Formal del Mercado Objetivo mediante CIIU

Dado que el dataset del SRI clasifica actividades económicas mediante el código CIIU, se procede a identificar los códigos correspondientes a:

* Comercio al por menor (G47)

* Servicios de comida (I56)

* Servicios personales como peluquerías, servicio técnico (S96)

Este enfoque reduce sesgo y garantiza una segmentación objetiva.

### FILTRADO OFICIAL POR CIIU

```python

# FILTRADO OFICIAL POR CIIU
# NEGOCIOS DE BARRIO - CONECTAMIBARRIO


# Crear copia del dataset limpio
df_barrio = df.copy()

# Prefijos CIIU asociados a negocios de barrio
prefijos_barrio = (
    "G47",  # Comercio al por menor
    "I56",  # Restaurantes y comida
    "S96"   # Servicios personales (peluquerías, servicio técnico)
)

# Filtrar por prefijo del código CIIU
df_barrio = df_barrio[
    df_barrio["CODIGO_CIIU"].str.startswith(prefijos_barrio, na=False)
]

# Verificar tamaño del nuevo mercado objetivo
df_barrio.shape

# Este subconjunto representa comercios minoristas,
# locales de comida y servicios personales de barrio,
# alineados con el modelo de ConectaMiBarrio.
```

```python
df_barrio["CODIGO_CIIU"].str[:3].value_counts().head(15)
```

### CONTEO DE ESTABLECIMIENTOS POR CANTÓN

```python

# CONTEO DE ESTABLECIMIENTOS POR CANTÓN


conteo_canton = df_barrio["DESCRIPCION_CANTON_EST"].value_counts()

conteo_canton

```

### PORCENTAJE POR CANTÓN

```python

# PORCENTAJE DE PARTICIPACIÓN POR CANTÓN

# Calcular porcentaje respecto al total
porcentaje_canton = (
    conteo_canton / conteo_canton.sum() * 100
).round(2)

porcentaje_canton

```

## Identificación de Polos Comerciales en Pichincha

Con base en la concentración de establecimientos de barrio por cantón y parroquia, se identifican tres grandes polos comerciales dentro de la provincia de Pichincha.

Estos polos representan zonas estratégicas para el despliegue inicial del proyecto ConectaMiBarrio, debido a su alta densidad de comercio minorista y servicios de proximidad.

### Polo 1 — Quito Urbano Consolidado

Corresponde al núcleo comercial más fuerte de la provincia.
Incluye parroquias urbanas con alta densidad poblacional y elevada concentración de negocios de barrio como:

* Calderón (Carapungo)

* Iñaquito

* Chillogallo

* Cotocollao

* La Magdalena

* Centro Histórico

* Guamaní

* Quitumbe

* Kennedy

* Carcelén

Este polo concentra la mayor parte del mercado objetivo y representa el núcleo estratégico inicial.

### Polo 2 — Expansión Metropolitana

ncluye parroquias que forman parte del crecimiento urbano hacia los valles y zonas periféricas:

* Conocoto

* Sangolquí

* Tumbaco

* Cumbayá

Estas zonas presentan alta actividad comercial en expansión y potencial de digitalización progresiva.

### Polo 3 — Norte Provincial

Incluye cantones fuera del núcleo metropolitano pero con presencia comercial relevante:

* Cayambe

* Pedro Moncayo

* Pedro Vicente Maldonado

* San Miguel de los Bancos

* Puerto Quito

Representan mercados secundarios con menor densidad, pero estratégicos para expansión territorial futura.

## Clasificación de polos comerciales

```python
# 4 Clasificar parroquias en polos comerciales

# Crear columna de polo
df_barrio["POLO_COMERCIAL"] = "Otros"

# Polo 1 — Quito Urbano Consolidado
polo1 = [
    "CALDERON (CARAPUNGO)", "IÑAQUITO", "CHILLOGALLO",
    "COTOCOLLAO", "LA MAGDALENA", "CENTRO HISTÓRICO",
    "GUAMANÍ", "QUITUMBE", "KENNEDY", "CARCELÉN"
]

# Polo 2 — Expansión Metropolitana
polo2 = [
    "CONOCOTO", "SANGOLQUI", "TUMBACO", "CUMBAYA"
]

# Asignar polos
df_barrio.loc[
    df_barrio["DESCRIPCION_PARROQUIA_EST"].isin(polo1),
    "POLO_COMERCIAL"
] = "Polo 1 - Quito Urbano Consolidado"

df_barrio.loc[
    df_barrio["DESCRIPCION_PARROQUIA_EST"].isin(polo2),
    "POLO_COMERCIAL"
] = "Polo 2 - Expansion Metropolitana"

# Polo 3 se identifica por cantón fuera de Quito
df_barrio.loc[
    df_barrio["DESCRIPCION_CANTON_EST"].isin([
        "CAYAMBE", "PEDRO MONCAYO",
        "PEDRO VICENTE MALDONADO",
        "SAN MIGUEL DE LOS BANCOS",
        "PUERTO QUITO"
    ]),
    "POLO_COMERCIAL"
] = "Polo 3 - Norte Provincial"

df_barrio["POLO_COMERCIAL"].value_counts()
```

```python
# 7 Distribución porcentual por polo comercial

porcentaje_polos = (
    df_barrio["POLO_COMERCIAL"]
    .value_counts(normalize=True) * 100
).round(2)

import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
porcentaje_polos.plot(kind="bar")

plt.title("Distribución Porcentual de Negocios por Polo Comercial")
plt.xlabel("Polo Comercial")
plt.ylabel("Porcentaje (%)")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Este gráfico resume la concentración estratégica
# del mercado objetivo en la provincia.
```

## Visualización Geoespacial de Polos Comerciales

Para comprender la distribución territorial de los polos comerciales identificados, se procede a realizar una visualización geoespacial simplificada.

El objetivo es representar visualmente la concentración estratégica de negocios de barrio en la provincia de Pichincha, facilitando la toma de decisiones para la fase inicial del proyecto ConectaMiBarrio.

```python
conteo_polos = df_barrio["POLO_COMERCIAL"].value_counts()
conteo_polos.index
```

```python

# Visualización geoespacial de polos comerciales 

import folium
import numpy as np

mapa_polos = folium.Map(location=[-0.22, -78.50], zoom_start=10)

conteo_polos = df_barrio["POLO_COMERCIAL"].value_counts()

def agregar_polo(lat, lon, cantidad, nombre, color):
    folium.Circle(
        location=[lat, lon],
        radius=np.sqrt(cantidad) * 6,
        popup=f"{nombre}: {int(cantidad)} negocios",
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.5
    ).add_to(mapa_polos)

# -----------------------------
# Polo 1 — Quito Urbano Consolidado
# Norte (Carcelén / Calderón)
# Centro (Iñaquito / Centro Histórico)
# Sur (Quitumbe / Guamaní)
# -----------------------------

cantidad_polo1 = conteo_polos.get("Polo 1 - Quito Urbano Consolidado", 0)

agregar_polo(-0.10, -78.48, cantidad_polo1 * 0.33, "Polo 1 Norte", "blue")
agregar_polo(-0.20, -78.50, cantidad_polo1 * 0.34, "Polo 1 Centro", "blue")
agregar_polo(-0.30, -78.54, cantidad_polo1 * 0.33, "Polo 1 Sur (Quitumbe)", "blue")

# -----------------------------
# Polo 2 — Expansión Metropolitana
# Centrado correctamente en Sangolquí
# -----------------------------

cantidad_polo2 = conteo_polos.get("Polo 2 - Expansion Metropolitana", 0)

agregar_polo(-0.33, -78.45, cantidad_polo2, "Polo 2 Sangolquí", "green")

# -----------------------------
# Polo 3 — Norte Provincial (Cayambe zona referencial)
# -----------------------------

cantidad_polo3 = conteo_polos.get("Polo 3 - Norte Provincial", 0)

agregar_polo(0.05, -78.15, cantidad_polo3, "Polo 3 Norte Provincial", "red")

mapa_polos


# Agregar leyenda personalizada al mapa

legend_html = """
<div style="
position: fixed; 
bottom: 50px; left: 50px; width: 250px; height: 120px; 
background-color: white; 
border:2px solid grey; 
z-index:9999; 
font-size:14px;
padding:10px;
">
<b>Polos Comerciales</b><br>
<span style='color:blue;'>●</span> Polo 1 - Quito Urbano Consolidado<br>
<span style='color:green;'>●</span> Polo 2 - Expansión Metropolitana<br>
<span style='color:red;'>●</span> Polo 3 - Norte Provincial
</div>
"""

mapa_polos.get_root().html.add_child(folium.Element(legend_html))

mapa_polos

```
